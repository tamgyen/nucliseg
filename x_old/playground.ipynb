```python
from PIL import Image

# Open the PNG image using PIL
image = Image.open("C:/Dev/Projects/nucliseg/endometrium-dataset-analysis/data/dataset/dataset_coco_new_train/images/108.png")
image_array = np.array(image)

if image_array.shape[2] != 3:
    raise ValueError("The image should have 3 channels (RGB)")
print("Image shape:", image_array.shape)
```

```python
def plot_keypoints_on_image(image, keypoints, image_size=(256, 256)):
    fig, ax = plt.subplots()

    # Display the RGB image
    ax.imshow(image)

    for i in range(0, len(keypoints), 3):
        x, y, vis = keypoints[i], keypoints[i + 1], keypoints[i + 2]
        if vis:
            circle = plt.Circle((x, y), radius=2, color='r', fill=True)
            ax.add_patch(circle)

    ax.set_aspect('equal')
    ax.invert_yaxis()  # Invert y-axis to match image coordinates
    ax.set_xlim(0, image_size[0])  # Set x-axis limits
    ax.set_ylim(0, image_size[1])  # Set y-axis limits

    plt.show()

# Example RGB image (256x256)
image = np.random.rand(256, 256, 3)  # Replace this with your actual image data

# Example keypoints (x1, y1, vis1, x2, y2, vis2, ...)
keypoints = coconew["annotations"][0]["keypoints"]

plot_keypoints_on_image(image_array, keypoints, image_size=(256, 256))
```
